## Import Required Modules

```python
import numpy as np
import pandas as pd
import xarray as xa
import rioxarray as riox


import matplotlib.pyplot as plt
from matplotlib.dates import datestr2num, num2date
import seaborn as sns

from pathlib import Path
```

```python
sns.set_theme("talk")
sns.set_style("darkgrid")
```

## Define Functions

```python
def normalize(array):
    """Normalizes numpy arrays into scale 0.0 - 1.0"""
    array_min, array_max = array.min(), array.max()
    return ((array - array_min)/(array_max - array_min))

def make_composite(band_1, band_2, band_3):
    """Converts three raster bands into a composite image"""
    return normalize(np.dstack((band_1, band_2, band_3)))

def make_rgb_composite(raster):
    """Creates an RGB composite image using the input raster."""
    r = raster[2]
    g = raster[1]
    b = raster[0]

    return make_composite(r, g, b)

def calculate_ndvi(raster):
    """Calculates the normalised difference vegetation index using the input raster."""
    r = raster[2]
    nir = raster[3]

    return (nir - r) / (nir + r)

def calculate_ndwi(raster):
    """Calculates the McFeeters (1996) normalised difference water index using the input raster."""
    g = raster[1]
    nir = raster[3]

    return (g - nir) / (g + nir)

def extract_s2_ndvi_ndwi_timeseries(base_filepath):
    filepath_gen = Path(base_filepath).rglob('*/Sentinel_2_SR/*.tif')
    filepaths = [file for file in filepath_gen if file.is_file()]
    no_of_files = len(filepaths)

    # Sets up dummpy arrays/lists to be amended.
    median_ndvi = np.empty(no_of_files)
    median_ndwi = np.empty(no_of_files)
    date_list = []

    # Recursively iterating through each Sentinel 2 image in the base folder.
    for index, path in enumerate(filepaths):
        # Load in raster.
        raster = riox.open_rasterio(path)[:-1] / 10000
        # Load in cloud probability map.
        cloud_probability_map = riox.open_rasterio(path)[-1]
        # Masking out pixels with a 1 sigma detection of clouds.
        cloud_mask = cloud_probability_map < 68 
        
        # Extract date from filestring and append to list.
        date_list.append(f"{path.name[:4]}-{path.name[4:6]}-{path.name[6:8]}")

        # Calculate median NDVI and NDWI of cloud masked raster and 
        # add to relevant arrays.
        median_ndvi[index] = calculate_ndvi(
            raster.where(cloud_mask)
        ).median()
        
        median_ndwi[index] = calculate_ndwi(
            raster.where(cloud_mask)
        ).median()

    return date_list, median_ndvi, median_ndwi

def create_timeseries_df(date_list, median_ndvi, median_ndwi):
    df = pd.DataFrame(
        {
            "date": np.array(date_list),
            "datenum": datestr2num(date_list), # Converts dates to floats as some plotting functions do not accept str or datetime values.
            "ndvi": median_ndvi,
            "ndwi": median_ndwi,
        }
    )

    return df
```

## NDVI and NDWI Time Series Data

Here I am attempting to extract NDVI and NDWI timeseries data from the Sentinel 2 images for the Shamshato Refugee Camp and Karachi. I then present timeseries regression plots showing the general trend of NDVI and NDWI over time, as well as showing a clear correlation between the chosen NDVI and NDWI metrics.

#### Extracting the Time Series Data

```python
%%time

date_list_shamshato, median_ndvi_shamshato, median_ndwi_shamshato = extract_s2_ndvi_ndwi_timeseries('/content/drive/Shareddrives/ICG Data Analytics/Data/Pakistan/Shamshato')
```

```python
%%time

date_list_karachi, median_ndvi_karachi, median_ndwi_karachi = extract_s2_ndvi_ndwi_timeseries('/content/drive/Shareddrives/ICG Data Analytics/Data/Pakistan/Karachi')
```

##### Insert extracted data into a Pandas DataFrame


```python
shamshato_df = create_timeseries_df(date_list_shamshato, median_ndvi_shamshato, median_ndwi_shamshato)
karachi_df = create_timeseries_df(date_list_karachi, median_ndvi_karachi, median_ndwi_karachi)

display(shamshato_df.head(), karachi_df.head())
```

### Plotting Time-Series Regression Plots using Seaborn

#### Defining a function to convert numerical dates into string format for plot tick labels

```python
@plt.FuncFormatter
def fake_dates(x, pos):
    """ Custom formater to turn floats into e.g., 2016-05-08"""
    return num2date(x).strftime('%Y-%m-%d')
```

#### Shamshato

##### NDVI Timeseries

```python
fig, ax = plt.subplots(figsize=(15,8))
sns.regplot(x='datenum', y='ndvi', data=shamshato_df, ax=ax)
ax.xaxis.set_major_formatter(fake_dates)
ax.tick_params(labelrotation=45)
plt.xlabel("Date")
plt.ylabel("NDVI")
plt.title("Shamshato Refugee Camp")
plt.savefig("/content/drive/Shareddrives/ICG Data Analytics/Events/2021-11-17 Innovation Monthly Meeting/Plots/Shamshato_Median_NDVI_Timeseries_Regression_MaxFM.png", bbox_inches="tight")
```
