## Importing raster data using rioxarray

The image bands are (in order) VV and VH. 




```python
s1_filepath = "/content/drive/Shared drives/ICG Data Analytics/Data/Pakistan/Karachi/2020-08-01/Sentinel_1_SAR_GRD_C_BAND/SAR_Karachi_20200828T012629.tif"

# Bands are (in order) VV and VH. No cloud probability is given here as the bands are IR and so see through clouds.

s1_raster = riox.open_rasterio(s1_filepath) # Scaled VV and VH bands

s1_raster
```

### Make RGB composite, NDVI and NDWI images

```python
vv = s1_raster[0]
vh = s1_raster[1]
```

## Plot static images with matplotlib

```python
plt.subplots(figsize=(15,15))
plt.imshow(vv)
```

```python
plt.subplots(figsize=(15,15))
plt.imshow(vh)
```
